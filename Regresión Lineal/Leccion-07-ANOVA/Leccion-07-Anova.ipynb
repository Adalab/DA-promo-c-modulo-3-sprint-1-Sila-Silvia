De los tres dataframes que tenemos (normalizado, estandarizado con el método Standard y estandarizado con el método Robust) hemos decidido hacer el estudio con el dataframe estandarizado con el método Standard ya que hasta que no hagamos la regesión lineal no podemos llegar a conclusiones fiables.

### ANOVA del dataframe estandarizado con el método Standard

```python
# Iniciamos nuestro anova. Observamos que nuestras variables predictoras son de efecto no aditivos, es decir,
# entendemos que no se influyen mutuamente (son independientes) a la hora de tomar la decisión de hacer la compra online.
lm = ols('numwebpurchases_sklearn ~ id + year_birth + education + marital_status + income + dt_customer + numdealspurchases + numstorepurchases + numwebvisitsmonth' , data=df_standard).fit()
sm.stats.anova_lm(lm)
```

Cuando realizamos nuestra anova en el dataframe estandarizado con el método "Standard" vemos que los grados de libertad en las columnas categóricas es en education "4.0" (hay 5 valores únicos), marital_status "7.0" (hay 8 valores únicos), dt_customer "661.0" (hay 662 valores únicos). En el caso de las variables numéricas vemos que el resultado de todas es 1, ya que solo pueden variar dentro de un eje.

Un dato que nos llama la atención es el valor de dt_customer (661.0).

Analizando el PR(<F), es decir el "p-valor" debemos tener en cuenta que nuestra hipótesis nula es que las variables predictoras no influyen en la compra online.
Teniendo en cuenta esto observamos que "education" tiene un valor mayor que 0.05, por lo que no influye en la variable respuesta, es decir, en las compras online. Todo el resto de las variables tiene un valor menor a 0.05, por lo que si influyen en la variable respuesta.

Una vez hecho este estudio lo confirmamos a través de un summary.

```python
lm.summary()
```

Analizando el "summary" lo primero que observamos es el R2 (tiene un valor de 0.678). Al ser el valor del R2 elevado sabemos que hemos utilizado unas variables predictoras para nuestro modelo correctas, ya que el cambio de nuestra variable respuesta casi un 70% depende de nuestras variables predictoras.

Observamos a la hora de realizar el "summary" que en las variables predictoras id y year_birth el p-value es completamente diferente (sus valores en uno y otro estudio nos reflejan respuestas contradictorias). Nos quedaremos con el p-value que nos aparece en el summary. En este caso el valor es superior a 0.05, por lo tanto no tiene efecto en la variable respuesta.

En cuanto al parámetro "coef" vemos que tanto "income", "numdealspurchases", "numstorepurchases" y "numwebvisitsmonth" tienen resultados positivos por lo que influye directamente en la compra online, es decir, a mayor ingresos, ofertas, compras en tiendas y número de visitas, mayor es el número de compras online.

En conclusión, vemos que los principales valores que influyen en la compra online según nuestro estudio son: "income", "numdealspurchases", "numstorepurchases", "numwebvisitsmonth", "marital_status" y "dt_customer" no influyendo todos por igual en la variable respuesta (lo que nos puede llevar a eliminar alguna variable predictora en un futuro). 

Hemos decidido eliminar dt_customer del estudio por su alto número de valores únicos.

```python
df_standard.drop(["year_birth", "id", "education","dt_customer"], axis=1, inplace=True)
```
