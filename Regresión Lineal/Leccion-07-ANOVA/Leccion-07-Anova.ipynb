### ANOVA del dataframe estandarizado con el m茅todo Standard

```python
# Iniciamos nuestro anova. Observamos que nuestras variables predictoras son de efecto no aditivos, es decir,
# entendemos que no se influyen mutuamente (son independientes) a la hora de tomar la decisi贸n de hacer la compra online.
lm = ols('numwebpurchases_sklearn ~ id + year_birth + education + marital_status + income + dt_customer + numdealspurchases + numstorepurchases + numwebvisitsmonth' , data=df_standard).fit()
sm.stats.anova_lm(lm)
```

Cuando realizamos nuestra anova en el dataframe estandarizado con el m茅todo "Standard" vemos que los grados de libertad en las columnas categ贸ricas es en education "4.0" (hay 5 valores 煤nicos), marital_status "7.0" (hay 8 valores 煤nicos), dt_customer "661.0" (hay 662 valores 煤nicos). En el caso de las variables num茅ricas vemos que el resultado de todas es 1, ya que solo pueden variar dentro de un eje.

Un dato que nos llama la atenci贸n es el valor de dt_customer (661.0).

Analizando el PR(<F), es decir el "p-valor" debemos tener en cuenta que nuestra hip贸tesis nula es que las variables predictoras no influyen en la compra online.
Teniendo en cuenta esto observamos que "education" tiene un valor mayor que 0.05, por lo que no influye en la variable respuesta, es decir, en las compras online. Todo el resto de las variables tiene un valor menor a 0.05, por lo que si influyen en la variable respuesta.

Una vez hecho este estudio lo confirmamos a trav茅s de un summary.

```python
lm.summary()
```

Analizando el "summary" lo primero que observamos es el R2 (tiene un valor de 0.678). Al ser el valor del R2 elevado sabemos que hemos utilizado unas variables predictoras para nuestro modelo correctas, ya que el cambio de nuestra variable respuesta casi un 70% depende de nuestras variables predictoras.

Observamos a la hora de realizar el "summary" que en las variables predictoras id y year_birth el p-value es completamente diferente (sus valores en uno y otro estudio nos reflejan respuestas contradictorias). Nos quedaremos con el p-value que nos aparece en el summary. En este caso el valor es superior a 0.05, por lo tanto no tiene efecto en la variable respuesta.

En cuanto al par谩metro "coef" vemos que tanto "income", "numdealspurchases", "numstorepurchases" y "numwebvisitsmonth" tienen resultados positivos por lo que influye directamente en la compra online, es decir, a mayor ingresos, ofertas, compras en tiendas y n煤mero de visitas, mayor es el n煤mero de compras online.

En conclusi贸n, vemos que los principales valores que influyen en la compra online seg煤n nuestro estudio son: "income", "numdealspurchases", "numstorepurchases", "numwebvisitsmonth", "marital_status" y "dt_customer" no influyendo todos por igual en la variable respuesta (lo que nos puede llevar a eliminar alguna variable predictora en un futuro). 

Hemos decidido eliminar dt_customer del estudio por su alto n煤mero de valores 煤nicos.

```python
df_standard.drop(["year_birth", "id", "education","dt_customer"], axis=1, inplace=True)
```
